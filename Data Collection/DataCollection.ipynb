```python
import requests
import time
import json
import pandas as pd
import random

rito_key = ''
```

```python
# api url to get information for my account
api_url_account = 'https://americas.api.riotgames.com/riot/account/v1/accounts/by-riot-id/Jubjub/9727?api_key=' + rito_key

# get api request for my account
account_request = requests.get(api_url_account)

if account_request.status_code == 200:
  # if request succeeded print the response and save my personal puuid as a variable
    print(account_request.json())
    puuid = account_request.json()['puuid']
else:
  # else print the status code
  print('Error: ', account_request.status_code)
```

```python
game_history_api_url = 'https://americas.api.riotgames.com/lol/match/v5/matches/by-puuid/' + puuid + '/ids?type=normal&start=0&count=75&api_key=' + rito_key

# get api request for my game history
game_history_reuqest = requests.get(game_history_api_url)

if game_history_reuqest.status_code == 200:
  # if request succeeded print the response and save my personal puuid as a variable
    print(game_history_reuqest.json())
    game_history = game_history_reuqest.json()
else:
  # else print the status code
  print('Error: ', game_history_reuqest.status_code)
```

```python
unique_puuid = set()
for match_id in game_history:

  match_api_url = 'https://americas.api.riotgames.com/lol/match/v5/matches/' + match_id + '?api_key=' + rito_key

  # get api request for match
  match_reuqest = requests.get(match_api_url)

  if match_reuqest.status_code == 200:
    # if request succeeded print the response and save my personal puuid as a variable
      for pluuid in match_reuqest.json()['metadata']['participants']:
        unique_puuid.add(pluuid)
  else:
    # else print the status code
    print('Error: ', match_reuqest.status_code)
```

```python
def get_50_matches(puuid):
  game_history_api_url = 'https://americas.api.riotgames.com/lol/match/v5/matches/by-puuid/' + puuid + '/ids?type=normal&start=0&count=50&api_key=' + rito_key

  # get api request for my game history
  while True:
    game_history_reuqest = requests.get(game_history_api_url)

    if game_history_reuqest.status_code == 429:
      print('Rate limit exceeded. Waiting for 1 minute...')
      time.sleep(60)
      continue
    else:
      break

  if game_history_reuqest.status_code == 200:
    # if request succeeded print the response and save my personal puuid as a variable
      return game_history_reuqest.json()
  else:
    # else print the status code
    print('Error: ', game_history_reuqest.status_code)
    print(game_history_reuqest.headers)
```

```python
unique_matches = set()

for puuid in unique_puuid:
    matches = get_50_matches(puuid)

    for match in matches:
          unique_matches.add(match)
```

```python
def get_match(match_id):
  match_api_url = 'https://americas.api.riotgames.com/lol/match/v5/matches/' + match_id + '?api_key=' + rito_key

  while True:
    match_reuqest = requests.get(match_api_url)

    if match_reuqest.status_code == 429:
      print('Rate limit exceeded. Waiting for 1 minute...')
      time.sleep(60)
      continue
    else:
      break

  if match_reuqest.status_code == 200:
    # if request succeeded print the response and save my personal puuid as a variable
      return match_reuqest.json()
  else:
    # else print the status code
    print('Error: ', match_reuqest.status_code)
```

```python
limited_set = random.sample(unique_matches, 10000)
```

```python
data_list = []

for match_id in limited_set:
    match = get_match(match_id)
    data_list.append (match)
```

```python
with open('match_data.json', 'w') as file:
    json.dump(data_list, file, indent=4)
```
